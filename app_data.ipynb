Next we want to separate the genres. This will make it easier for use to use one hot encoder to have multiple values for genres. We are going to split the entry by '&' and ';'

```python
app.Genres = app.Genres.str.replace(" ", "")
temp = app.Genres.str.split(';|&').apply(pd.Series)
print(temp)
```

```python
df = (pd.get_dummies(temp[0])
        .add(pd.get_dummies(temp[1]), fill_value=0)
        .add(pd.get_dummies(temp[2]), fill_value=0)
        .add(pd.get_dummies(temp[3]), fill_value=0)
        .astype(int))

df1 = pd.DataFrame(df)
df1[['Art','Design','Education','Creativity','Medical','PretendPlay']]
```

```python
app = app.drop('Genres',axis=1)
app = pd.concat([app, df1],axis=1)
```

Now we want to add 'CATEGORY' to the one hot encoder but we need to do the same as Genres. We only want to replace '0' the missing genres that change it to'1'. We first want to isolate categories to another dataframe to do the manipulations

```python
temp = pd.DataFrame(app.Category).convert_dtypes()
temp.info()
```

We first want to format by lowercasing it and then uppercasing the title

```python
temp.Category = temp.Category.str.replace('_', ' ')
temp.Category = temp.Category.str.title()
temp.Category = temp.Category.str.replace(' And ', ';')
temp.Category = temp.Category.str.replace(' ', '')

temp.Category.unique()
```

This table 

```python
temp = temp.Category.str.split(';').apply(pd.Series)
print(temp)
```

```python
cate_df = (pd.get_dummies(temp[0])
        .add(pd.get_dummies(temp[1]), fill_value=0)
        .astype(int))

cate_df = pd.DataFrame(cate_df)
cate_df
```

To have only '1' be checked off we will have all '0' turn into Nan then replace those only with the second table.

```python
app[app.columns[12:]] = app[app.columns[12:]].replace(0, np.nan)
app[app.columns[12:]]
```

We will add the missing columns from app to cate_df and reindex the columns to be the same format alphabetically

```python
for x in app.columns[12:]:
    if not x in cate_df.columns:
        cate_df[x] = 0

cate_df = cate_df.reindex(sorted(cate_df.columns), axis=1)
cate_df
```

```python
for i in cate_df.columns:
    if not i in app.columns:
        app[i] = np.nan
        
range = app.columns[12:]
app[range]
```

```python
app[range] = app.reindex(sorted(app[range]), axis=1)
app[range]
```

Now we will add col to cate_df to to be able to merge to app

```python
app[range] = app[range].combine_first(cate_df)
app[range].loc[app['Family'] == 1.0]
app = app.drop('Category', 1)
```

### Convert installs into something suitable for the computer

The first step is to convert the numbers into strings and get rid of the + and convert into float64

```python
app.Installs = app.Installs.str.replace('+', '')
app.Installs.value_counts()
```

```python
app.Installs = app.Installs.str.replace('Free', '0')
app.Installs = app.Installs.str.replace(',', '')
app.Installs.unique()
```

```python
app.Installs = pd.to_numeric(app.Installs)
app.Installs
```

## Linear Regression

## Split the data

```python
from sklearn.model_selection import train_test_split
range = app.columns[11:]
X, y = app[range], app.Installs
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression

my_lin = LinearRegression()

my_lin.fit(X_train, y_train)
```

```python
print(my_lin.intercept_)
print(my_lin.coef_)
```

As you can see the accuracy is really low so we will normalize the y will help us make the accuracy better

```python
y_pred = my_lin.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

m = metrics.mean_squared_error(y_test, y_pred)
r = np.sqrt(m)
print(r)
```

## Applying 10-fold Cross Validation for 'Linear Regression'

```python
from sklearn.model_selection import cross_val_score

mse_list = cross_val_score(my_lin, X, y, cv=76, scoring='neg_mean_squared_error')
print(mse_list)
```

```python
mse_list_pos = -mse_list

r = np.sqrt(mse_list_pos)
print(r)
```
